```python
#contour
xlist = np.linspace(-3.0,3.0,100) #100개의 데이터 
ylist = np.linspace(-3.0,3.0,100)
x,y = np.meshgrid(xlist, ylist) #두개씩 짝을지어 좌표 짝을지우는 명령 -> 10000개의 접점이 만들어진다.
#mesh = 그물망 gird = 격자 , 그물망 격자
Z = np.sqrt(x**2+y**2) #100개의 데이터
fig,ax = plt.subplots(1,1)
cp = ax.contourf(x,y,Z) #등고선 -> 2차원표현 (데이터는 3차원이 필요하다.)
fig.colorbar(cp)
ax.set_title('contour')
ax.set_ylabel('y(cm)')
plt.show()
```

```python
from mpl_toolkits import mplot3d #projection='3d', ax.plot3D 를 사용할 수 있는 확장팩
fig = plt.figure()
#도화지를 3차원으로 획득
ax = plt.axes(projection='3d') #프로젝션 : polar 원형좌표계에 출력하는것을 의미
z = np.linspace(0,1,100)
x = z*np.sin(20*z) #5바퀴정도
y = z*np.cos(20*z)
ax.plot3D(y,x,z,'gray')
ax.set_title('3d_plot')
plt.show()
#각이지는 이유는 듬성듬성 결과가 있기 때문이다.
#원일 수가 없다.
#갖고있는 기능 anti-aliasing으로 부드럽게 표현이 된다.

#원으로 돌때 좌표값(cos,sin)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
x, y = np.meshgrid(X,Y)
R = np.sqrt(x**2+y**2) #제곱한 값을 싸인함수에 전달(값이 커지더라고 일정한 주기를 생성하게 되어진다.)
Z = np.sin(R*2) #값에 *곱하기를 해주면 더 빨리 각도가 움직이게 된다.
ax.plot_surface(x,y,Z, rstride=1, cstride=1, cmap='hot') #면을 그린다,출력한다.
#sutface는 원래 삼각형으로 면을 형성한다.(부드러운 면을 형성하기 위해서)
#위에서 봤을때 격자모양을 준거고 높이값만 준것이다.
plt.show()
ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=plt.cm.rainbow) #파도를 정면에서 보는것처럼 높은점과 낮은점을 반복해서 울렁이게된다.
plt.show()
```

```python
def f(x,y): return(1-x/2+x**5+y**3) * np.exp(-x**2-y**2) #5승이면 변곡점이 4개 있어야하낟.
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
x,y = np.meshgrid(X,Y)
Z = np.sin(f(x,y))
ax.plot_surface(x,y,Z, rstride=1, cstride=1, cmap='hot') #격자를 1로줘라
ax.view_init(elev=50, azim=200) #보는 관점을 바꿔준다.
plt.show()
```

```python
#f : fill 채워라 채워서 표시해라
plt.contourf(x,y,f(x,y), 8, alpha=.75, cmap='jet') #8 - 얼마나 세밀하게 표시할 것인가?
#오차방정식을 놓고 높이값을 sin값을 주었다. (변곡점4개 생성)
plt.show() #파란색은 낮고 빨간색은 높다.
```
