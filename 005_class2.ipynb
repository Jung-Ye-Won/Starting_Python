```python
import sys
class Student: #반복하는 문제고 1인분에 대해서 정의를 잘 해주면 된다.
    def __init__(self):
        self.bunho=0
        self.name=''
        self.kor=0
        self.mat=0
        self.eng=0
        self.total=0
        self.average=0
        self.grade=''
    def inputData(self):
        self.name=input("이름을 입력하시오")
        self.kor=eval(input("국어점수"))
        self.mat=eval(input("수학점수"))
        self.eng=eval(input("영어점수"))
    def calc_total_average(self):#선언만 되어있는 변수에 값을 대입
        self.total=self.kor + self.mat + self.eng 
        self.average=round(self.total/3,2)
        self.grade=self.calcSemGrade()
    def calcSemGrade(self):
        if self.average>=90 : return 'A'
        elif self.average>=80 : return 'B'
        elif self.average>=70 : return 'C'
        elif self.average>=60 : return 'D'
        else : return 'F'
    def __repr__(self):
        return '%5s %5s %6.2f %6.2f %6.2f %7.2f %7.2f %5s' %(
            self.bunho, self.name, self.kor, self.mat, self.eng, self.total, self.average, self.grade)
```

```python
class FourCal: #__init__없는데 멤버변수 생성가능
    #자바나 C에서는 생성자 아닌곳에서 변수생성이 불가능 했었다.
    def setdata(self, first, second):
        self.first=first #self=멤버변수
        self.second=second #어떠한 멤버함수도 멤버변수를 생성할 수 있다.
    def sum(self):
        result=self.first + self.second
        return result
    def mul(self):
        result = self.first * self.second
        return result
    def sub(self):
        result = self.first - self.second
        return result
    def div(self):
        result = self.first/self.second
        return result
    def all_cal(self):
        result1 = self.sum()
        result2 = self.mul()
        result3 = self.sub()
        result4 = self.div()
        return result1, result2, result3, result4
fourcal = FourCal()
fourcal.setdata(100,200)
fourcal.first=200 #멤버변수처럼 사용가능
fourcal.second=300
fourcal.third=500 #실시간으로 객체에 변수와 함수를 추가가 가능하다.
print(fourcal.third)

def addnum(a,b):
    return a+b
fourcal.addnum = addnum #함수를 변수처럼 넘겨주었다.
print(fourcal.addnum(10,20)) #실시간으로 함수를 추가가 가능하다. (전부 포인터에 의해서 작동하기 때문에 가능하다.)
fourcal.all_cal()
```

```python
class Guard:
    def three_point(self):
        print("가드는 3점슛 장착")
    def apass(self):
        print("가드의 핵심은 패스")
class Jordan:
    def three_point(self):
        print("조단은 3점슛 장착")
    def apass(self):
        print("조단은 에어패스 가능")
#delegation(대리자)
#기본적으로 파이썬베이스에서 제공하는 함수 : 실수를 하지 않으려면 상속을 받아야한다.
#상속구조를 띄어야지 객체에 없는 함수를 호출하는 경우가 없어 실수하지 않을 수 있다.
def inhecourt(duck): #출력을 공통으로 진행하는 함수
    duck.three_point(); duck.apass()
gu = Guard()
jo = Jordan()

inhecourt(gu) #함수를 각각호출하는 대신 대리자개념으로 매개변수를 객체로 받아 함수를 호출할 수 있게끔 해준다.
inhecourt(jo)
```

```python
#문제 : Point class와 shape class를 만들고 shape를 상속받은 triangle, rectangle, circle class를 생성하시오
#면적과 둘레를 구하는 함수는 공용입니다.
#각 도형 클래스 10개씩을 만들고 이를 실행시켜 보시오
#mapping을 잘해야한다.
import math

class Point: 
    def __init__(self, x, y):
        self.x = x
        self.y = y
class Shape:
    def area(self):
        pass
    def circumference(self):
        pass
class Triangle(Shape): 
    def __init__(self, point1, point2, point3):
        self.point1 = point1
        self.point2 = point2
        self.point3 = point3
    def circumference(self): #둘레
        dist1 = math.sqrt((self.point1.x-self.point2.x)**2 + (self.point1.y-self.point2.y)**2) 
        dist2 = math.sqrt((self.point3.x-self.point2.x)**2 + (self.point3.y-self.point2.y)**2) 
        dist3 = math.sqrt((self.point1.x-self.point3.x)**2 + (self.point1.y-self.point3.y)**2)
        cir = round(dist1+dist2+dist3)
        return "삼각형의 둘레는 " + str(cir)
    def area(self):
        rec_area = 0.5*((self.point1.y-self.point2.y) * 
                        (self.point3.x-self.point2.x)) - ((self.point1.x-self.point2.x) * (self.point3.y-self.point2.y))
        return "삼각형의 넓이는 " + str(round(rec_area,2))
class Rectangle(Shape): 
    def __init__(self, point1, point2):
        self.point1 = point1
        self.point2 = point2
    def circumference(self): #둘레
        cir = abs(self.point1.x-self.point2.x)*2 + abs(self.point1.y-self.point2.y)*2
        return "사각형의 둘레는 " + str(cir)
    def area(self):
        rec_area = abs(self.point1.x-self.point2.x) * abs(self.point1.y-self.point2.y)
        return "사각형의 넓이는 " + str(rec_area)
class Circle(Shape): 
    def __init__(self, center, rad):
        self.center = center
        self.rad = rad
    def circumference(self): #둘레
        peri = math.pi*self.rad*2
        return "원의 둘레는 " + str(peri)
    def area(self):
        cirarea = math.pi * float(self.rad**2)
        return "원의 넓이는 " + str(cirarea)
    
class Shape_print:
    def print_shape(Self, evaluation):
        print(evaluation.circumference())
        print(evaluation.area())
```

```python
point1 = Point(10,3)
point2 = Point(5,20)
point3 = Point(7,8)

shapes = [Triangle(point1,point2,point3), Rectangle(point1, point2), Circle(point2,5)]
test = Shape_print()
for a in shapes: 
    test.print_shape(a)
    
import random
shape_list=[]
for i in range(100):
    rand = random.randint(0,10)
    if rand%3 ==0:
        tri = Triangle(Point(20,30), Point(20,10), Point(10,50))
        shape_list.append(tri)
    elif rand%3 ==1:
        rect = Rectangle(Point(10,20), Point(20,30))
        shape_list.append(rect)
    else:
        circle = Circle(Point(10,20),5)
        shape_list.append(circle)
for i in range(100):
    print(shape_list[i].area())
    print(shape_list[i].circumference())
```

```python
#초침에 의해서 움직이는 시계
class Clock(object): #지역변수를 상속이 불가능하다.
    def __init__(self, hours=0, mins=0, sec=0): #초기값0으로 설정
        self.__hours=hours #private 변수
        self.__mins=mins
        self.__sec=sec
    def set(self, hours, mins, sec): #강제로 바꾸고 싶을때 사용하라고
        self.__hours=hours 
        self.__mins=mins
        self.__sec=sec
    def tick(self):
        if self.__sec==59:
            self.__sec=0
            if self.__mins ==59:
                self.__mins=0
                if self.__hours==23:
                    self.__hours=0
                else:
                    self.__hours+=1
            else:
                self.__mins+=1
        else:
            self.__sec+=1 #돌아가면서 이게 증가되어서 숫자가 올라가는것이다.
    def Checkhours(self):
        if self.__sec==0 and self.__mins==0 and self.__hours==0:
            return 1;
        else: return 0;
    def display(self):
        print("%d:%d:%d"%(self.__hours, self.__mins, self.__sec))
    def __str__(self):
        return "%d:%d:%d"%(self.__hours, self.__mins, self.__sec)
```

```python
#날짜에 의해서 움직이는 달력
class Calender(object):
    months=[31,28,31,30,31,30,31,31,30,31,30,31]
    def __init__(self, day=1, month=1, year=1900):
        self.__day = day
        self.__month = month
        self.__year = year
    def leapyear(Self, y): #윤년계산(2월이 29일이 된다.)
        if y%4:
            return 0 #윤년아니다.
        else:
            if y%100:
                return 1 #윤년이다.
            else:
                if y%400:
                    return 0 #윤년이아니다.
                else:
                    return 1 #윤년이아니다.
    def set(self, day, month, year):
        self.__day=day
        self.__month=month
        self.__year=year
    def advance(self):
        months = Calender.months
        max_day=months[self.__month-1]
        if self.__month ==2:
            max_day+self.leapyear(self.__year) #윤년일 경우 1더해준다.
        if self.__day ==max_day: #월이변경되야한다.
            self.__day =1
            if self.__month ==12:
                self.__month=1
                self.__year+=1 #12월일 경우 년도수정
            else:
                self.__month+=1
        else:
            self.__day+=1
    def __str__(self):
        return str(self.__year)+"/"+str(self.__month)+"/"+str(self.__day)
```
